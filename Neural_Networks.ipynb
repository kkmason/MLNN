### 1. Using the XOR dataset below, train (400 epochs) a neural network (NN) using 1, 2, 3, 4, and 5 hidden layers (where each layer has only 2 neurons). For each n layers, store the resulting loss score along with n. 

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
plt.scatter(*zip(*X), c=y)
```

```python
type(y)
```

### 2 Neurons - 1 Hidden Layer

```python
# because the output must be equal to y, units must be 1. 

model = Sequential()

model.add(Dense(1, input_dim=2, activation='tanh'))  #sigmoid, relu
model.summary()

sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer='sgd')

model.fit(X, y, batch_size=2, epochs=400) #160/4 = 40 per epoch
print(model.predict_proba(X).reshape(4*n))

# evaluate the model
scores = model.evaluate(X, y)
```

```python
loss_scores = []

loss_scores.append({'layers':1,'loss_score':scores, 'metric':model.metrics_names})
loss_scores
```

```python
plt.scatter(*zip(*X), c=model.predict_classes(X))
```

```python
plt.scatter(*zip(*X), c=model.predict(X))
```

### 3. Using the most optimal configuraion (n-layers, k-neurons per layer), compare how `tanh`, `sigmoid`,`softplus` and `relu` effect the loss after 400 epochs. Try other Activation functions as well (https://keras.io/activations/)

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD  #Stochastic Gradient Descent

import numpy as np
# fix random seed for reproducibility
np.random.seed(7)

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
n = 40
xx = np.random.random((n,1))
yy = np.random.random((n,1))
```

```python
X = np.array([np.array([xx,-xx,-xx,xx]),np.array([yy,-yy,yy,-yy])]).reshape(2,4*n).T
y = np.array([np.ones([2*n]),np.zeros([2*n])]).reshape(4*n)
```

```python
# for loop:

loss_scores = []
activations = ['tanh','sigmoid','relu', 'hard_sigmoid','softsign', 'linear', 'softplus','selu']
# model = Sequential()

for activation in activations:
    
    model = Sequential()
    model.add(Dense(4, input_dim=2, activation=activation))  #sigmoid, relu
    model.add(Dense(4, activation=activation))
    model.add(Dense(4, activation=activation))
    model.add(Dense(1, activation='sigmoid'))


    sgd = SGD(lr=0.1)
    model.compile(loss='binary_crossentropy', optimizer='sgd')
    # model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])

    model.fit(X, y, batch_size=2, epochs=400) #160/4 = 40 per epoch
    print(model.predict_proba(X).reshape(4*n))

    # evaluate the model
    scores = model.evaluate(X, y)

    loss_scores.append({'activation': activation,'loss_score':scores, 'metric':model.metrics_names})

    print(loss_scores)
```

```python
loss_scores = pd.DataFrame(loss_scores)
loss_scores
```

```python
loss_scores.plot(x='model',kind='bar')
```

#### Looking at the bar chart, tanh activation achieved the lowest loss score of all models, with relu an softsign not far behind. The loss score for both sigmoid and hard_sigmoid models did not improve over the iterations of epochs.

### 4. Instead of `SGD` try other optimizers and report on the loss score. (https://keras.io/optimizers/)

```python
# from keras import optimizers
from keras.optimizers import SGD
from keras.optimizers import RMSprop
from keras.optimizers import Adagrad
from keras.optimizers import Adadelta
from keras.optimizers import Adam
from keras.optimizers import Adamax
from keras.optimizers import Nadam
```

```python
loss_scores = []
optimizers = ['sgd','rmsprop','adagrad','adadelta','adam','adamax','nadam']


for optimizer in optimizers:
    
    model = Sequential()

    model.add(Dense(4, input_dim=22, activation='tanh'))  #sigmoid, relu
    model.add(Dense(4, activation='tanh'))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(loss='binary_crossentropy', optimizer=optimizer)


    model.fit(X, y, batch_size=200, epochs=400) #160/4 = 40 per epoch

    # evaluate the model
    scores = model.evaluate(X, y)
    # scores

    loss_scores.append({'optimizer': optimizer,'loss_score':scores, 'metric':model.metrics_names})

    print(loss_scores)
```
